# Local application of functions to lists

```python
def square(x):
    return x * x
```

```python
parameters = [0, 1, 2, 3, 4, 5]
```

```python
[square(x) for x in parameters]
```

```python
mapped = map(square, parameters)
```

```python
mapped
```

```python
list(mapped)
```

# Remote application of functions to lists with pywren

```python
import pywren

pwex = pywren.default_executor()
```

```python
futures = pwex.map(square, parameters)
```

```python
futures
```

```python
[f.result() for f in futures]
```
